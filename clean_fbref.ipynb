```python
import pandas as pd
```

```python
df_defense = pd.read_csv("raw/fbref_defense.csv")
df_goal_shot_creation = pd.read_csv("raw/fbref_goal_shot_creation.csv")
df_misc = pd.read_csv("raw/fbref_misc.csv")
df_passing_types = pd.read_csv("raw/fbref_passing_types.csv")
df_playing_time = pd.read_csv("raw/fbref_playing_time.csv")
df_possession = pd.read_csv("raw/fbref_possession.csv")
df_shooting = pd.read_csv("raw/fbref_shooting.csv")
df_standard = pd.read_csv("raw/fbref_standard.csv")

df_keeper = pd.read_csv("raw/fbref_keeper.csv")
df_keeper_adv = pd.read_csv("raw/fbref_keeper_adv.csv")
```

```python
def rename_column(column_name: str) -> str:
    return column_name.replace(" ", "_").lower()

df_defense = df_defense.rename(mapper=rename_column, axis=1)
df_goal_shot_creation = df_goal_shot_creation.rename(mapper=rename_column, axis=1)
df_misc = df_misc.rename(mapper=rename_column, axis=1)
df_passing_types = df_passing_types.rename(mapper=rename_column, axis=1)
df_playing_time = df_playing_time.rename(mapper=rename_column, axis=1)
df_possession = df_possession.rename(mapper=rename_column, axis=1)
df_shooting = df_shooting.rename(mapper=rename_column, axis=1)
df_standard = df_standard.rename(mapper=rename_column, axis=1)
df_keeper = df_keeper.rename(mapper=rename_column, axis=1)
df_keeper_adv = df_keeper_adv.rename(mapper=rename_column, axis=1)
```

```python
df_keeper_all = pd.merge(
    left=df_keeper,
    right=df_keeper_adv[[
        "player",
        "team",
        "goals_fk",
        "goals_ck",
        "goals_og",
        "expected_psxg",
        "expected_psxg/sot",
        "expected_psxg+/-",
        "expected_/90",
        "launched_cmp",
        "launched_att",
        "launched_cmp%",
        "passes_att_(gk)",
        "passes_thr",
        "passes_launch%",
        "passes_avglen",
        "goal_kicks_att",
        "goal_kicks_launch%",
        "goal_kicks_avglen",
        "crosses_opp",
        "crosses_stp",
        "crosses_stp%",
        "sweeper_#opa",
        "sweeper_#opa/90",
        "sweeper_avgdist",
    ]],
    on=["player", "team"],
    suffixes=(None, "_adv")
)

df_keeper_all.to_csv("clean/fbref_keeper.csv", index=False)
```

```python
df_general_all = pd.merge(
    left=df_standard,
    right=df_defense[[
        "player",
        "team",
        "tackles_tkl",
        "tackles_tklw",
        "tackles_def_3rd",
        "tackles_mid_3rd",
        "tackles_att_3rd",
        "challenges_tkl",
        "challenges_att",
        "challenges_tkl%",
        "challenges_lost",
        "blocks_blocks",
        "blocks_sh",
        "blocks_pass",
        "int",
        "tkl+int",
        "clr",
        "err"
    ]],
    on=["player", "team"],
)
```
